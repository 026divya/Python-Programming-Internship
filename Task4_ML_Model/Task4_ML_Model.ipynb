```python
import pandas as pd

data = {
    "message": [
        "Win money now",
        "Hello friend",
        "Free gift offer",
        "Meeting tomorrow",
        "Earn cash fast",
        "Let's have lunch",
        "Congratulations you won",
        "Project discussion today"
    ],

    "label": [
        "spam",
        "ham",
        "spam",
        "ham",
        "spam",
        "ham",
        "spam",
        "ham"
    ]
}

df = pd.DataFrame(data)

df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(df["message"])

y = df["label"]

print("Text converted into numbers!")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)

print("Data split successfully!")
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()

model.fit(X_train, y_train)

print("Model trained successfully!")
```

```python
predictions = model.predict(X_test)

print("Predictions:")
print(predictions)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)

print("Model Accuracy:", accuracy)
```

```python
sample_message = ["Free money offer"]

sample_vector = vectorizer.transform(sample_message)

result = model.predict(sample_vector)

print("Message:", sample_message[0])
print("Prediction:", result[0])
```
